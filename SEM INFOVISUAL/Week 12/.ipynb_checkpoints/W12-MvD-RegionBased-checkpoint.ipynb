# <font color =  #3498db> 1- Region-Based Plotting Techniques : Multivariate Data using superimposed multiple displays

### 1.1-  Using Multi-Bar chart  (aka Clustered bar chart)

We can plot the values of multiple variable as __clustered bars__ on the same diaplay. Each cluster consists of a bar for each variable. Usually each variable is assigned a distinct color to identify it in the legnd. The clusters are labeled with values of a categorical variable or index, and the each cluster's label is usually placed at the center of the cluster. 

In the following example we will plot `Var1,Var2,Var3,Vr4` values as bar charts on the same dispaly (figure) and we will label each cluster by month name

```python
#
plt.figure(figsize=(15,6))
x_axis = np.arange(len(d))

# How many data variables we are going to plot as bars, i.e. the number of bars in each cluster
numberOfVariables = 4

# bar width is calculated based on the number of bars in each cluster +1 for the gab
w = 1.0/(numberOfVariables+1) 


#plot the 1st variable (Var1) values as bars in the first position of each cluster, and use color green for these bars
#The first postion of the cluster is the index we generated as x_axis above
var1= plt.bar(x = x_axis ,width=w,height= d["Var1"], color="green",label="Var1",align='edge')

#plot the 2nd variable (Var2) values as bars in the 2nd position of each cluster, and use color blue for these bars
# the second position starts from where the privious bar ends, i.e. at x_axis+ bar width
var2= plt.bar(x = x_axis+w ,width=w,height= d["Var2"], color="b",label="Var2",align='edge')

#plot the 3rd  variable (Var3) values as bars in the 3rd position of each cluster, and use color red for these bars
# the third position starts after the privious two bars end, i.e. at x_axis+ two bar widths
var1= plt.bar(x = x_axis+w*2 ,width=w,height= d["Var3"], color="r",label="Var3",align='edge')

#smae as previous, color cyan
var1= plt.bar(x = x_axis+w*3 ,width=w,height= d["Var4"], color="c",label="Var4",align='edge')


#To position the x-tixks at the center of each cluster we calculate the center based on the number of 
#      bars (variables) and the width of each bar
clusterCenter = w * numberOfVariables/2
tick_locations = x_axis+  clusterCenter #Instead, we can center it in cluster space (including the gab) by adding 0.5 (x_axis+0.5)

plt.xticks(tick_locations, d["Month"], rotation=90)

# Create a legend for our chart
plt.legend(loc="best")

# Show the chart
plt.show()


#NOTE: The calculations of the bar and xticks positions based on the alignment of the bars at the edge (align='edge')
#       if we used "default" center alignment (align='center') calculations of xticks positions neeed to be adjusted

#Warning: The *left* kwarg to `bar` is deprecated use *x* instead.
#var1= plt.bar(left = x_axis ,width=w,height= d["Var1"], color="green",label="Var1")
```

### 1.2- Using Stacked bars 

Use cases for stacked vertical bar charts are more limited. They are
meant to allow you to compare totals across categories and also see
the subcomponent pieces within a given category

In the following example we will plot `Var1,Var2,Var3,Vr4` values as stacked bars on the same dispaly (figure) and we will label each stack by month name. The plot will allow us compare our variables by category (month) also, to some extent, compare the variations in each variable and each category.   

```python
# Plot the "Var1" data
plt.figure(figsize=(15,6))
x_axis = np.arange(len(d))
w=0.4 # here we will have fixed bar width

#plot the 1st variable (Var1) values as bars in the first position of each stack, and use color green for these bars
#The first postion of the stack is at the bottom = None (default)
var1= plt.bar(x = x_axis,width=w,height= d["Var1"], color="green",label="Var1")

#plot the 2nd variable (Var2) values as bars in the 2nd position of each stack, and use color blue for these bars
# the second position starts from the privious variable's bar top, i.e. its bottom is the values of Var1, so bottom=d["Var1"]
var2= plt.bar(x = x_axis,width=w,height= d["Var2"],bottom= d["Var1"], color="b",label="Var2")

#plot the 3rd  variable (Var3) values as bars in the 3rd position of each stack, and use color red for these bars
# the third position starts after the privious two bars, i.e. its bottom is at the position (d["Var1"] + d["Var2"]) 
var3= plt.bar(x = x_axis,width=w,height= d["Var3"],bottom= d["Var1"]+d["Var2"], color="r",label="Var3")

#same as previous, with bottom at  d["Var1"]+d["Var2"]+d["Var3"] and use cyan color
var4= plt.bar(x = x_axis,width=w,height= d["Var4"], bottom= d["Var1"]+d["Var2"]+d["Var3"],color="c",label="Var4")


tick_locations = x_axis  # since we use the default alignment "center"

plt.xticks(tick_locations, d["Month"], rotation=45)

# Create a legend for our chart
plt.legend(loc="best")

# Show the chart
plt.show()
```

### 1.3 -  Using Percentages stacked bars

Percentages stacked bars is a special case of stacked bars, but we use percentages of the variable values instead of the values. The stacks will have the same length, however the regions inside the stack differ based on the percentages

##### Example: 
To plot the previous varaibles using percentages stacked bars, We first need to compute the contribution percentage of each variable in each month, so we will calculate total of each month (row) then devide each variable's value in each row by the total of its row (month). 

* __Step1:__ calculate the percentages

```python
# calculate the sum of each row execluding the 1s element (which is the month) 
monthsTotals = d.iloc[:,1:].sum(axis=1)  
monthsTotals
```

```python
# Devide each row element by its row total
percentages = d.iloc[:,1:].divide(monthsTotals, axis=0) *100
#percentages.sum(axis=1)
percentages
```

To check your calculation (sanity check) the sum of each row should be 100

```python
#check, sum columns in each row
percentages.sum(axis=1)
```

Or you can do the same thing using loops if you like

```python
#get the values in a new dataframe
percentages = d.iloc[:,1:]

#update the values in each cell by deviding it by its row (month) total
for i in range(0, len(d)):
    for j in range(0,4):
        percentages.iloc[i,j]= percentages.iloc[i][j]/monthsTotals[i] * 100
```

```python
percentages.sum(axis=1)
```

* __Setp2:__ Plot the percentages using the same procedure we used to plot stacked bars

```python
# Plot the "Var1" data
plt.figure(figsize=(15,6))
x_axis = np.arange(len(d))
w=0.6

totals = d.iloc[:,1:].sum(axis=1)

percentages = d.iloc[:,1:].divide(totals, axis=0) *100

var1= plt.bar(x = x_axis,width=w,height= percentages["Var1"], color="green",label="Var1")
var2= plt.bar(x = x_axis,width=w,height= percentages["Var2"],bottom= percentages["Var1"], color="b",label="Var2")
var3= plt.bar(x = x_axis,width=w,height= percentages["Var3"],bottom= percentages["Var1"]+percentages["Var2"], color="r",label="Var3")
var4= plt.bar(x = x_axis,width=w,height= percentages["Var4"], bottom= percentages["Var1"]+percentages["Var2"]+percentages["Var3"],color="c",label="Var4")


tick_locations = x_axis 

plt.xticks(tick_locations, d["Month"], rotation=45)

plt.ylabel('Percentages %')
# Create a legend for our chart
plt.legend(bbox_to_anchor = (1,1), loc="upper left")

# Show the chart
plt.show()
```

### 1.4- Using Nested pie (donut) charts

Nested Pie plot is useful to show the relationship between variables when one of the variables is composed by the other variable, for example aggregation (like sum). The following example shows `Var1` and its Quarters totals (to be calculated) relationship, where Quarters is composed by aggregating `Var1` values by year quarters (3 months intervals). 

* __step1:__ calculate the four quarters totals of `Var1`

```python
# Let's calculate the quarters by summing every three consective months
# To do that, get the values of var1 as array, then reshape it as x by 3 then sum across the columns
#  Note: The -1 in the 1st dimension in reshape means we don't know (or we don't care) what would be 
#        the size of that dimension, so reshape() will calculate it. in this case will be 4, since we have 12 values 

Quarters = d['Var1'].values.reshape(-1, 3).sum(axis=1) 
 # you cna also do it using Quarters = d['Var1'].values.reshape(4, 3).sum(axis=1) 
Quarters
```

* __step2:__ plot 2 superimposed donuts, one for the Quartes and one for `Var1` values

```python
fig, ax = plt.subplots(figsize=(8,8))

size = 0.3
#Normalize values to map colors
normV1 = (1-0.1) *(d["Var1"]-d['Var1'].min())/(d['Var1'].max()-d['Var1'].min()) + 0.1
normQs = (1-0.1) *(Quarters-Quarters.min())/(Quarters.max()-Quarters.min()) + 0.1

#Pick a color map
cmap = plt.get_cmap("Blues")

#Map colors to values
Q_colors = cmap(normQs) 
#Q_colors = ['#ff6666', '#ffcc99', '#99ff99', '#66b3ff']

V1_colors = cmap(normV1)

QLabels= ['Q1','Q2','Q3','Q4']

#Plot the outer donut for the Quarters
ax.pie(Quarters, labels=QLabels,
       labeldistance=1.1,             # put the Qs labels outside, note distance>1
       explode=[0.01]*len(Quarters),  # To increase the gab between wedges, note it takes a list with size = length of the values
                                       #explode specifies as a list the fraction of the radius with which to offset each wedge
                                       # multiplying the list by scaller repeats the values in the list, so we get [0.01,0.01,0.01,0.01]
       textprops={'color':"yellowgreen",'fontsize':14}, # set the label text properties color and fontsize
       autopct='%1.1f%%', pctdistance=0.85,  # add percentages to the wedges at 0.8 distance from the center
       radius=1,                  # set the percentage of the radious of the outer dounut, 1 means all space (100%)
       colors=Q_colors,           # use the mapped colors to color the wedges
       wedgeprops=dict(width=size, edgecolor='w') # set wedge properties width and edge color, width <1 means donut
        )  

#Plot the inner donut for the months, note the radius=1-size to squeez it inside the outer donut
ax.pie(d['Var1'],labels=d["Month"],
       labeldistance=0.7, 
       rotatelabels=True,   #  to rotate labels along the wedge angle 
       textprops={'color':"yellow",'fontsize':12},
       #autopct='%1.1f%%', pctdistance=0.6,       # we dont want %s in the inner donut, uncomment if you want to add them
       radius=1-size,   # to squeez it inside the outer donut set the radius less than the width of the outer donut's wedges 
       colors=V1_colors,
       wedgeprops=dict(width=size, edgecolor='w'))  

ax.set(aspect="equal", title='Var1 vs. Its quarters')
plt.show()
```

# <font color =  #3498db> 2- Region-Based Plotting Techniques : Multivariate Data using Juxtaposed multiple displays

Lets plot the four variables (`Var1, Var2, Var3 and Var4`) in juxtaposed displays (subplots)

#### Ex1: Using 2x2 Grid with shared x-axis and y-axis labels

```python



figure1 = plt.figure(figsize = (8,4))

x_axis = np.arange(len(d))

figIdx=0
maxy = max(d[d.columns[1:]].max())
miny = min(d[d.columns[1:]].min())
colors = ['g','b','r','c']
w = 0.6
for column in d.columns[1:]:
    figIdx+=1
 
    # Find the right spot on the plot
    plt.subplot(2,2, figIdx)
 
    # Plot the bar
    var1= plt.bar(x = x_axis ,width=w,height= d[column], color=colors[figIdx-1],label="Var1",align='center')
    
    
    #plt.xlim(0,11)
    plt.ylim(miny,maxy)
    plt.tick_params(axis ='x' ,labelrotation =-45)
    tick_locations = x_axis 
    plt.xticks(tick_locations, d["Month"], rotation=45)
    #reove x label from top subplots
    if figIdx not in [3,4] :
        plt.tick_params(labelbottom=False)
     #reove y label from right subplots
    if figIdx not in [1,3] :
        plt.tick_params(labelleft=False)
 
    # Add title
    plt.title(column, loc='left', fontsize=12, fontweight=0, color='b' )
plt.tight_layout(pad=2.0) 
#General title,
# We use the argument y=1.05 to set the vertical location a little bet higer to leave space between the title and the plot
plt.suptitle("Example : bar plot for Var1, Var2, Var3 and Var4", fontsize=13, fontweight=0, color='black', style='italic', y=1.01)
 
plt.show()
```
