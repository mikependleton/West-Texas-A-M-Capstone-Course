In order to shed light on the subject of heart disease, we chose to look at and work with the chosen sutdy in order to understand the mortality rates of this sometimes invisable diease amongst the population, and to also see if location has any revelance to this problem.
This is intended for any audiance with an interest towards the same subject, and is meant to increase the total knowledge of this disease, without bias agianst or towards any particular group.

We chose a bar plot in order to visualize the data, finding that it fit best for the dataset.  Though it does have some troubles, it makes distinct counts for individual states, which other such plots would not have done nearly as well.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
MedData=pd.read_csv('Heart_Disease_Mortality_Data_Among_US_Adults__35___by_State_Territory_and_County.csv')
```

```python
MedData[:5]
```

```python
CleanMD=MedData[["LocationAbbr","Data_Value"]]
CleanMD.isnull().sum()
```

```python
CleanMD=CleanMD.dropna()
CleanMD['Data_Value']=CleanMD['Data_Value'].astype(int)
CleanMD.reset_index()
CleanMD[:5]
```

```python
CleanMD.describe()
```

The data shown here is is interesting to be certain, with very notable traits; though it is invisable in the visualization below, as it was impossible to use the original dataset.  Noteworthily, the min is 66 deaths per 100,000, while the max is 3000 out of 100,000; both huge outliers considering the mean and standard deviation.  Otherwise, heart disease can be said to take about .5% of the population, at least during this chosen year (2014)

```python
Data=CleanMD.groupby(['LocationAbbr']).mean()
Data['Data_Value']=Data['Data_Value'].astype(int)
Data
```

```python
CleanMD=CleanMD.dropna()
CleanMD['Data_Value']=CleanMD['Data_Value'].astype(int)
CleanMD.reset_index()
Data=CleanMD.groupby(['LocationAbbr']).mean()
Data['Data_Value']=Data['Data_Value'].astype(int)
```

```python
Data.plot.bar(figsize=(20,10))
plt.show()
```

We wondered how differences in location might have an effect on heart disease mortality data.  The chosen visualization clearly naswers this question; yes, it appears to be related, though it appears clear that location is but a single part of a much larger puzzle.

Data variance shows several intriguing maxes and minimums.  Guam shows the highest rate of heart disease, while the virgin isles shows the least, respectively averaging  529 and 88.  Large states have lower death counts than smaller ones (California has 297 deaths which is on the low end of the average), and it seems land bound states have higer counts that costal ones at a brief glance.

More variables would better the data, such as age of patient, sex, or economic stance, but that data was unavailable.  What we have does appear to show coorolation with location, with surprising numbers in the whole data as well as in the average.
Overall, we feel that this shines good light on the subject at hand, that being the count of deaths by heart disease by location, and hope that this will better increase the knowledge of the subject.
