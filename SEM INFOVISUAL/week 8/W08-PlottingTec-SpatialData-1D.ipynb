## <font color=blue> 1-Visualization Techniques for Spatial Data </font>

Spatial data has an intuitive, and often straightforward,
mapping of the data attributes to graphical attributes of the entities
conveying the information in the visualization.

To create a visualization of spatial data:
* Select the spatial attributes of the data that will map to the spatial attributes (locations) on the screen.
* Select the other attributes of the data that you want to include in the visualization and map them to other components of the visualization, using the other visual variables we mentioned before such as color, texture, size or shape of a graphical entity.
* This will depend on the dimensionality of the data (the number of attributes/columns of the data).

### 1.1-Single-variable ( One-Dimensional ) Data

One-dimensional spatial data is often the result of accumulating samples or readings of some phenomenon while moving along a path in space (only one value per data item).


To visualize 1D spatial data:
* map the index (for example using order, like time) of spatial data to one of the screen dimensions and the data value itself to either the other screen dimension (for example to form a line graph, bar chart) or to the color/size of a mark or region along the spatial axis (for example to form a color bar, bie chart).

### 1.2-Multivariate Data:

If the data set is multivariate (it contains more than one variable or value per data entry) we still able to treat it using 1-D strategies using either juxtapositioning or superimpositioning.

* For line graphs, this would mean that either the visualization would consist of a stack of nonoverlapping graphs (graph containing plots of two or more variables with different line styles or colors to distinquish the data variables). 

* For bar graphs, we can also have multiple bars placed next to each other, or stacked on top of each other.

* We need to deal with the scale issues and convey the scale of each variable to the user.

In general, the following techniques can be used with multivariate data:

* ** dimension subsetting**: allowing the user to select a subset of the dimensions (variables/columns/attributes) to display

* **dimension reduction**: using techniques such as principal component analysis (PCA) or multidimensional scaling to transform the high-dimensional data to data of lower dimension, while attempting to preserve as best as possible the relationships among the data points

* ** dimension embedding** : mapping dimensions to other graphical attributes besides position, such as color, size, and shape (though there are limits to how many dimensions can be included this way).

* **multiple displays** : showing, either superimposed or juxtaposed, several plots, each of which contains some of the dimensions.



## <font color=blue> 2- Visualization techniques (tools)</font>


**Line Graphs:**

* line plots can be drawn on a common set of axes, differentiating the dimensions using color, line style, width, or other graphical attributes

* data: 2 quant attribs (one key, one value)
* marker: points and line
    * line connects markers 
* properties:
    * separated and ordered by key attrib into horizontal regions
* purpose
    * find trend
    * showing relationship between one item and the next
* scalability
    * hundreds of key levels, hundreds of value levels

* Limitations:
    * variables have different units
    * dimensions have significant overlap in their data ranges



**scatter plot:**
* data: quantitative attributes
    * require 2 quant attribs (no keys, only values)
* marker: points
* Main spatial property 
    * Position (horiz. + vert.) 
* purpose
    * find trends, 
    * show outliers, distribution, correlation, clusters
* scalability
    hundreds of items

**bar chart:**
* data: 
    * 1 categ attribute (index/key), 1 quantitative attrib. (value)
    
* marker: bar
* Properties:
    * length to express quant. attrib. (value) 
    * spatial regions: one per marker
    * if not ordered:
        * Hard to know rank. Whatâ€™s the 4th highest? The 5th?
    * ordered by quant. attrib.
        * by label (e.g. alphabetical), by length attrib (by value)
* purpose
    * compare values,
    * lookup values
* scalability
    * few levels for key attrib

**heatmap:**
* data: (two keys, one value)
    * 2 categ attribs
    * 1 quant attrib 
* markers: area
    * separate and align in 2D matrix
    * indexed by 2 categorical attributes
* properties:
    * color by quant. attrib.
   
* purpose
    * find clusters, outliers
* scalability
    * 1M items, 100s of categ levels, ~10 quant attrib levels

**parallel coordinates:**

* parallel axes, jagged line representing item (value as point on axis, line connect item points)
* major challenge: axis ordering 
* scalability
    * dozens of attribs (axes)
    * hundreds of items

**pie chart:**
* area markers with angle as encoded value attribute
* Limitations:
    * accuracy:
        * angle/area less accurate than line length
        * arclength also less accurate than line length

    
* data
    * 1 categ key attrib, 1 quant value attrib
* purpose
    * part-to-whole comparison
    
**polar area chart:**
* area markers with length variable
* analog to bar charts


**boxplot:**
* item aggregation
* purpose: find distribution
    
* data: 5 quant attribs (statistical)
    * median: central line
    * lower and upper quartile: boxes
    * lower upper limits: whiskers
        * values beyond which items are outliers
    

#### Selecting plot type:

Choosing the wrong visualization tool could cause confusion or lead to mistaken data interpretation.

There are many factors that we must consider when selecting a visual, starting from the reasons why we might need a chart (purpose), and the type and dimensionality of our data. 

We might need to ask :

* Are we comparing values?
* Do we want to show the composition of some variables or items in our data?
* Do we want to understand the distribution of your data?
* Are we interested in analyzing trends in the data? or the relationship between the data items?


For example:
* Line graphs are most commonly used to plot continuous data. Because the points are physically connected via the line, it implies a connection between the points that may not make sense for categorical data (a set of data that is sorted or divided into different categories). Often, our continuous data is in some unit of time: days, months, quarters, or years.

* Bar charts are easy for our eyes to read. Our eyes compare the end points of the bars, so it is easy to see quickly which category is the biggest, which is the smallest, and also the incremental difference between categories.

### 3.1- Using Line graph ``plt.plot()``

```python
# Plot the "Var1" data
var1, = plt.plot( d["Var1"], color="green",label="Var1")
x_axis = np.arange(len(d))
tick_locations = x_axis #[value for value in x_axis]

plt.xticks(tick_locations, d["Month"], rotation=90)

# Create a legend for our chart
plt.legend(loc="best")

# Show the chart
plt.show()
```

### 3.2- Using Bar chart  ``plt.bar()``

```python
# Plot the "Var1" data
x_axis = np.arange(len(d))

var1= plt.bar(x_axis , d["Var1"], color="green",label="Var1")

tick_locations = x_axis #[value for value in x_axis]

plt.xticks(tick_locations, d["Month"], rotation=90)

# Create a legend for our chart
#plt.legend(loc="best")

# Show the chart
plt.show()
```

**Changing the colors of the bars**

**Step 1**: Select the colors (color map)

```python
# Create a color map 
cmap = plt.get_cmap("gist_yarg")  # if size is omitted, takes the default value
cmap.N
```

**Step 2:**  Map the values to the colors:

One way to make the selected colors cover the range of the selected color map is to  normalize the data.
here we are going to normalize the data to be between 0.1 and 1  to avoid those hard to see colors

We can use the following formula to normalize data X to be between the values a and b 

${\displaystyle X'=\left(b-a\right) {\frac {\left(X-X_{\min }\right)}{X_{\max }-X_{\min }}} + a}$

```python
#Let's map the values to the colors:

# normalize the data values 1st
normalizedValues = (1-0.1) *(d["Var1"]-d['Var1'].min())/(d['Var1'].max()-d['Var1'].min()) + 0.1

# map the colors from the color map to the values
colors = cmap(normalizedValues)
```

```python
# Plot the quant. variable "Var1" as data values vs the categ. variable "Month" as index 
x_axis = np.arange(len(d))

#cmap = plt.get_cmap("gist_yarg")

#colors = cmap(d["Var1"]*10) 

# the following commented line will produce warning for **left**, use x instead 
#var1= plt.bar(left = x_axis ,height= d["Var1"], color=colors,label="Var1")

var1= plt.bar(x = x_axis ,height= d["Var1"], color=colors,label="Var1")

tick_locations = x_axis #[value for value in x_axis]

plt.xticks(tick_locations, d["Month"], rotation=90)

# Create a legend for our chart
#plt.legend(loc="best")

# Show the chart
plt.show()
```

### 1.3- Using Horizontal Bar chart ``plt.barh()``

### 3.4 - Using Pie Chart ``plt.pie()``

The unique thing you get with a pie chart is the concept of
there being a whole and, thus, parts of a whole. However, pie charts are not recommended by many authors because they are hard for people to read. When segments are close in size, itâ€™s difficult (if not impossible) to tell which is bigger. When they arenâ€™t
close in size, the best you can do is determine that one is bigger
than the other, but you canâ€™t judge by how much. One solution to this is to add percentages as labels.

    * The wedges are plotted counterclockwise, by default starting from the x-axis.

- Main parameters:
    *  x: The wedge sizes
    * explode : arry with which to offset each wedge.
    * labels : list contains a sequence of strings providing the labels for each wedge
    * colors : A sequence of matplotlib color through which the pie chart will cycle.  
    * autopct : If not *None*, is a string or function used to label the wedges with their numeric value.  The label will be placed inside the wedge.  If it is a format string, the label will be ``fmt%pct``.
    * pctdistance : float, default: 0.6,  The ratio between the center of each pie slice and the start of the text generated by *autopct*. 
    * labeldistance :  default: 1.1,  The radial distance at which the pie labels are drawn
    * wedgeprops : dictionary contains the properties and their values to apply for each  wedge 
    

```python
# Plot the "Var1" data as Pie chart

cmap = plt.get_cmap("Spectral")
colors = cmap(normalizedValues)
var1= plt.pie(d["Var1"],labels=d["Month"],colors=colors)

# Create a legend for our chart
#plt.legend(loc="best")

# Show the chart
plt.show()
```

It is important to preserve aspect ratio of the plot, so let's fix it using figure size.

```python
# Plot the "Var1" data as Pie chart
plt.figure(figsize=(7,7))
#plt.axes(aspect='equal')

var1= plt.pie( d["Var1"],labels=d["Month"],colors=colors)

# Create a legend for our chart
#plt.legend(loc="best")

# Show the chart
plt.show()
```

**Adding percentages to the pie chart**

Let's add percentages to the pie using ``autopct='%1.1f%%'`` , where the string '%1.1f%%' is formatting string to cotrol the number of dicimals displayed. In this case we want only one dicimal place.

```python
# Plot the "Var1" data as Pie chart
plt.figure(figsize=(7,7))
#plt.axes(aspect='equal')


# lebels are placed outside of the pie by default, to specify label location use labeldistance parameter
var1= plt.pie( d["Var1"],labels=d["Month"],colors=colors,autopct='%1.1f%%')


# Show the chart
plt.show()
```

### 3.5- Using Donut Chart  `plt.pie()`

Donut chart is a special case of Pie chart by changing the width property of the pie wedges.
* specify wedge width as a value between 0 (no wedges) and 1 (complete wedge)
* you can specify ``wedgeprops`` as a dictionary of key value pairs(``dict``) to set multiple properties of the wedge, like edgecolor, width,linewidth,.. etc.) 
* you can control the location of the label using the parameter ``labeldistance``, and the location of the percentage label using ``pctdistance``, which is the distance from the center. 0 at the center, 1 at the edges of the pie, >1 out of the pie area

```python

# Plot the "Var1" data as donut chart
plt.figure(figsize=(8,8))
#plt.axes(aspect='equal')

# note theat we used the last defined and mapped colors
size = .4
var1= plt.pie( d["Var1"],labels=d["Month"],labeldistance=1.03,
              autopct='%1.1f%%', pctdistance=.75,colors=colors,radius=1,
       wedgeprops=dict(width=size, edgecolor='w'))



# Show the chart
plt.show()
```

### 3.6 - Using Stacked bar  `plt.barh()`

For single variable (1-D) data, sometimes it is better to plot one horizontal stacked bars to give a sense of the subcomponent pieces. 
To do that, we need to know how:

    1- to plot horizontal bars
    2- iterate over the values and plot a bar for each value
    3-plot the next bar starting from where the previous bar ends

To add a legend out of the drawing area, we need to manually specify an anchor point that places the legend.
Specify the left lower corner and upper rihgt corner base ( [0,0] to  [1,1], -ve values or >1 vlues are out of the figure coordinates) cordinators using ``bbox_to_anchor`` parameter, and the location using the ``loc`` parametre by providing a string position name like 'lower left' or its code like ``loc=3``

See https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html for details and examples

```python
# Plot the "Var1" data
plt.figure(figsize=(10,0.5))
x_axis = np.arange(len(d))
left_pos = 0
for i in np.arange(len(d)):
    _= plt.barh(0, d.loc[i,"Var1"],height=0.5,edgecolor='black', left=left_pos, label=d.loc[i,"Month"])
    left_pos += d.loc[i,"Var1"]

#remove y-ticks
plt.yticks([])

# Adding Legend out of the plotting area (using -ve coordinates for anchor points)
anchor_vals = (0, -2, 1, -0.5)
plt.legend(bbox_to_anchor=anchor_vals, 
          loc=3, 
          ncol=6,  # number of columns in the legend table
          mode="expand", 
          borderaxespad=0.0)
#remove axis
#plt.axis('off')
plt.show()
```

## 4- Legends

### 4.1-Creating a legend for a plot
There are three ways to create a legend for your plot, as follows:

##### 1- Using Automatic detection of elements to be shown in the legend:
 * call `plt.legend()` without any parameters and use labels with the plotting functions
    

##### Example:

```python
plt.plot([1, 2, 3], label='Inline label1')
plt.plot([3, 2, 1], label="Inline label2")
plt.legend() # just call legend()
plt.show()
```

#### 2-Use `plt.legnd()` to add labels to already plotted elements
    
   * simply call `plt.legend()`  with a list of strings, one for each legend item.

##### Example:

```python
plt.plot([1, 2, 3])
plt.plot([3, 2, 1])
plt.legend(['line1','line2'])
plt.show()
```

#### 3- Use plot element handels and labels in `plt.legend()`
* pass a list of handels followed by a list of legend labels respectively
    
#### Example:

```python
line1, = plt.plot([1, 2, 3])
line2, = plt.plot([3, 2, 1])
line3, = plt.plot([2, 2, 2],linestyle='dashed')

plt.legend([line1, line2, line3], ['line1 label', 'line2 label', 'line3 label'])
plt.show()
```

### 4.2-Positioning the legend

There are also multiple ways to position the legend. The easiest way is to use the parameter `loc='best'` to automatically put the legend in the best place by matplotlib. However sometimes you want to manually control the position of the legend, for example to position it outside the figure boarders.

The follwoing figure shows all possible values for `loc` parameter and their locations
<table><tr><td><img src="LegendMapLoc.bmp"  width="450"></td><td><img src="LegendLoc.png"  width="250"></td></tr></table>


* In addition we can use `bbox_to_anchor`  in conjunction with `loc` to position the legend in arbitrary locations, even outside the boarder. Note that the figure box abstarct coordinates ar between `(0,0)` and `(1,1)`, any value outside this range will lace the legend outside the figure. The follwoing are the coordinates of the main locations :

    * 'C': (0.5, 0.5),
    * 'E': (1.0, 0.5), 
    * 'N': (0.5, 1.0), 
    * 'NE': (1.0, 1.0),
    * 'NW': (0, 1.0), 
    * 'S': (0.5, 0),
    * 'SE': (1.0, 0),
    * 'SW': (0, 0), 
    * 'W': (0, 0.5)}


* `bbox_to_anchor` could be a 2-tuple, or 4-tuple of floats. 
     * If a 4-tuple is given, then it specifies the bbox (x, y, width, height) that the legend is placed in. 
         * For example: to put the legend in the best location in the bottom right quadrant of the axes (or figure):
                `loc='best', bbox_to_anchor=(0.5, 0., 0.5, 0.5)`

     * A 2-tuple (x, y) places the corner of the legend specified by loc at x, y.
         * For example, to put the legend's upper right-hand corner in the center of the axes (or figure) use the following parameters:
                `loc='upper right', bbox_to_anchor=(0.5, 0.5)`
