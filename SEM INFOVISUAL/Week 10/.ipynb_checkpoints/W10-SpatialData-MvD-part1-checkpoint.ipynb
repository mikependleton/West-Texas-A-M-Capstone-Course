##  <font color =  #34981c> Plottin Multivariate Data using Scatter plots (Point-Based Technique)

* Data Dimension is the number of variables/attributes/columns the data has.
* Point-Based Techniques rely on our innate abilities to judge relative position within a bounded space (2-D or 3-D space). 
* As the dimensionality of the data to be visualized increases, we may need to use additional techniques like:
    * <b> dimension subsetting:</b> allowing the user (or algorithm) to select a subset of the dimensions to display
    * <b> dimension embedding:</b> mapping dimensions to other graphical attributes besides position, such as color, size, and shape (though there are limits to how many dimensions can be included this way).
    * <b> multiple displays:</b> showing, either superimposed or juxtaposed, several plots, each of which contains some of the dimensions.

# <font color =  #3498db> 1- Scatter plot of  Multivariate Data using dimension embedding

Scatterplots can be useful for showing the relationship between two variables, because they allow you to encode data simultaneously on a horizontal x‐axis and vertical y‐axis to see whether and what relationship exists. It is also possible to incorporate more than 2 variables in scatter plot by embedding the other variables (dimension) using the techniques we discussed in previous lectures, like marker size, color, and shape.

<b> Dimension embedding: </b> is mapping dimensions to other graphical attributes besides position, such as color, size, and shape (though there are limits to how many dimensions can be included this way).

 Our Example dataset is a multivariate Data with multiple (4) quantitative variables, let's see how we can plot this data using scatter plot 
 

## <font color =  #34911b> Example 1: Plotting Two  Quantitative variables

* #### Example 1.1: Scatter plot of `Var1` against `Var2`
In this example, we will plot the quantitative variables `Var1` aginst `Var2`, so we will use position as our visual variable by assigning `var1` values to x-axis and `var2` variable tp y-axis.

<img src="Inkeddata-anv1-v2_LI.jpg" width="200" height="100">

```python
# Var1 vs Var2
figure1 = plt.figure(figsize = (10,5))
_=plt.scatter(d["Var1"],d["Var2"],  alpha=0.8,cmap='winter')
plt.title("Example : 2 Quantitative Vars Scatter plot")
plt.xlabel("Var1")
plt.ylabel("Var2")
plt.show()
```

* #### Example1.2: Scatter plot of `Var3` against `Var4`

```python
# Var3 vs Var4
figure1 = plt.figure(figsize = (10,5))
_=plt.scatter(d["Var3"],d["Var4"],  alpha=0.8,cmap='winter')
plt.title("Example : 2 Quantitative Vars Scatter plot")
plt.xlabel("Var3")
plt.ylabel("Var4")
plt.show()
```

## <font color =  #349811> Example 2: Plotting Three Quantitative variables 

* ### <font color =  #1418db> Example 2.1:  Using Size to embed the 3rd variable:

Since for 2-D plots position can be identified using two variables (x,y) we need to use dimension embedding technique to map the 3rd variable to the properties of the marker (point) like color, shape, or size. Since we don't have too many shapes that can be easily distinguished in this example we will use `size`. That is:

* ``Var1`` Vs ``Var2`` as main axes (x,y) , and ``Var4`` as **size** of marker (dot)
* We will scale the variable `Var4` in order to better see the effect of the size. (for example to have sizes range between 1-100) 

<img src="InkedInkeddata-anv1-v2-v4-size.jpg" width="200" height="100"> 

```python
# Var1 Vs Var2 as main axes, and Var3 as size of markers
figure1 = plt.figure(figsize = (10,5))
sizes =(100-1) *(d["Var4"]-d["Var4"].min()/(d["Var4"].max()-d["Var4"].min())) +1

_=plt.scatter(d["Var1"],d["Var2"], s=sizes, alpha=0.8,cmap='winter')
plt.title("Example : 3 Quantitative Variables (Var1 vs Var2 and Var4 as Size) Scatter plot (bubble chart)")
plt.xlabel("Var1")
plt.ylabel("Var2")
plt.show()
```

* ### <font color =  #1418db> Example 2.2: Using Color to embed the 3rd variable: 

* ``Var1`` Vs ``Var2`` as main axes, and ``Var3`` as **Color** of markers

```python
# Var1 Vs Var2 as main axes, and Var4 as Color of markers
figure1 = plt.figure(figsize = (10,5))

colors = d["Var3"]
_=plt.scatter(d["Var1"],d["Var2"], c= colors, alpha=0.8,s = 100, cmap='winter')
plt.title("Example : 3 Quantitative Variables (Var1 vs Var2 and Var3 as color) Scatter plot (bubble chart)")
plt.xlabel("Var1")
plt.ylabel("Var2")
plt.show()
```

## <font color =  #349811> Example 3 : Plotting Four Quantitative variables

* ``Var1`` Vs ``Var2`` as main axes, and ``Var4`` as **size** of marker, ``Var3`` as **Color** of marker

```python
# Var1 Vs Var2 as main axes, and Var3 as size of marker, Var4 as Color of marker
figure1 = plt.figure(figsize = (10,5))
sizes =(100-1) *(d["Var4"]-d["Var4"].min()/(d["Var4"].max()-d["Var4"].min())) +1

#sizes = (d["Var3"]-d["Var3"].min()) * 20
colors = d["Var3"]
_=plt.scatter(d["Var1"],d["Var2"], c= colors,s=sizes, alpha=0.8,cmap='winter')
plt.title("Example : 4 Quantitative Variables (Var1 vs Var2, Var 3 as color and Var4 as Size) \n Scatter plot (bubble chart)")
plt.xlabel("Var1")
plt.ylabel("Var2")
plt.show()
```

We can change other attributes (mark shape, style,..)  of the visual to add more variables.

# <font color =  #3498db> 2- Scatter Plot of  Multivariate Data using superimposed multiple displays

In this type of plots each of the variables (dimension) is plotted separately against some index or variable on the same plot (superimposed), and usually each variables has unique marker shape or color.

## <font color =  #349811>  Example:  Plotting Four Quantitative variables using superimposed displays

* We will use index (or month) as x-axis, and plot each of our four variables ( `Var1`, `Var2`,`Var3`, and `Var4`) as y-axis using different colors on the same display (figure)

<img src="data-idx-v1v2v3v4.png" width="200" height="100">

```python
# Var1 Var2, Var3 and  Var4 as y-values on the same display (figure)
figure1 = plt.figure(figsize = (10,5))

v1=plt.scatter(d["Month"],d["Var1"], c= 'blue',s=50, marker='*' , alpha=0.8)
v2=plt.scatter(d["Month"],d["Var2"], c= 'red',s=50, marker = 'o', alpha=0.8)
v3=plt.scatter(d["Month"],d["Var3"], c= 'green',s=50,marker = '.', alpha=0.8)
v4=plt.scatter(d["Month"],d["Var4"], c= 'black',s=50,marker='+', alpha=0.8)

plt.title("Example : 4 Quantitative Vars Scatter plot ")
plt.xlabel("Month")
plt.ylabel("Var1, Var2, Var3, Var4")
plt.legend()
plt.show()
```

## <font color =  #3498db>  Making figures with multiple subplots 

<div class="alert alert-block  alert-info" > 
   matplotlib has many functions to add subplots to your figure where you can have multiple plots arranged in your prefered layout. `plt.subplots()` creates a figure and a grid of subplots with a single call, while  `plt.subplot()` adds one subplot with every call. See the following matplot   
    <a "href=https://matplotlib.org/3.1.0/gallery/subplots_axes_and_figures/subplots_demo.html">documentation link !</a> for more details .
    
</div>

* We usually call `plt.subplots()` as follows:

`fig , ax = plt.subplots(nRows,nCols)`

Where the parameters `nRows,nCols` specify the subplots layout (how many rows and columns). 
The returned values are figure object `fig`,  and  `ax` which can be either a single `~matplotlib.axes.Axes` object or an
array of Axes objects if more than one subplot was created.

* The other method `plt.subplot()` sdds a subplot to the current figure and it is called as follows:

` subplotAxes = plt.subplot(nRows,nCols, index)`

Where the parameters` nRows,nCols` specify the subplots layout (how many rows and columns). The `index` parameter specify the position on a grid with `nrows x ncols`. `index` starts at 1 in the upper left corner and increases to the right.

<div class="alert alert-block  alert-warning" >
 Note the slight differences between `plt.subplots()` and `plt.subplot()`. It is not just the extra `s` in the method name
    </div>

### <font color =  #1418db> Following are some examples form the documentation.

```python
#First create some toy data:
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)
```

 ### <font color =  #3498db> 1- Examples using `plt.subplots()`

```python
 #Creates two subplots as a grid of one row x two columns

f, ax = plt.subplots(1, 2, sharey=True)  # Sharey =True to make subplots to share y-axis
f.set_figheight(5)
f.set_figwidth(15)
f.suptitle("X vs. $sin(x^2)$")
ax[0].plot(x, y)
ax[0].set_title('Sharing Y axis')
ax[1].scatter(x, y)
plt.show()
```

```python
  #Creates four SUBPLOTS, and accesses them use TWO of them through the returned axes array
fig, mysubplots = plt.subplots(2, 2)
mysubplots[0, 0].plot(x, y)
mysubplots[1, 0].scatter(x, y)
plt.show()
```

```python
#Share a X axis with each column of subplots
plt.subplots(2, 2, sharex='col')
plt.show()
```

```python
#Share a Y axis with each row of subplots
plt.subplots(2, 2, sharey='row')
plt.show()
```

```python
#Share both X and Y axes with all subplots
plt.subplots(2, 2, sharex='all', sharey='all')
plt.show()
```

```python
# this is the same as above
plt.subplots(2, 2, sharex=True, sharey=True)
plt.show()
```

###  <font color =  #3498db> 2-Examples using `plt.subplot()`

```python
# making 2 x2 grid of subplot by adding each subplot separately

subplt1 = plt.subplot(2, 2,1)
subplt1.plot(x, y)

subplt3 = plt.subplot(2, 2,2)
subplt3.plot(x, y, c='red')
subplt4 = plt.subplot(2, 2,2)
subplt4.scatter(x, y,marker = '+', c='g')
plt.show()
```

#  <font color =  #3498db> 3-Scatter Plot of  Multivariate Data using juxtaposed multiple displays

We can utilize the matplotlib subplots to make juxtaposed figures of multiple variables.

### <font color =  green> Example:  scatter plots of  Four Quantitative variables using juxtaposed displays

* We will use index (or month) as x-axis, and plot each of our four variables ( `Var1`, `Var2`,`Var3`, and `Var4`) as y-axis using different subplot for each variable 

<img src="data-idx-v1v2v3v4.png" width="200" height="100">

#### Method 1: Using `plt.subplot()`

```python
# Var1 Var2, Var3 and  Var4 as y-values \
figure1 = plt.figure(figsize = (10,5))



var1plot = plt.subplot(2, 2,1)
v1=var1plot.scatter(d["Month"],d["Var1"])
var1plot.set_title("Var1")
var2plot = plt.subplot(2, 2,2)
v2=var2plot.scatter(d["Month"],d["Var2"])
var2plot.set_title("Var2")
var3plot = plt.subplot(2, 2,3)
var4plot = plt.subplot(2, 2,4)

#var2plot.title = "Example : 4 Quantitative Vars Scatter plot "

v3=var3plot.scatter(d["Month"],d["Var3"])
var3plot.set_title("Var3")

v4=var4plot.scatter(d["Month"],d["Var4"])
var4plot.set_title("Var4")


# adjusting the space between the subplots
figure1.tight_layout(pad=2.0) 

# we can use the following too
plt.subplots_adjust(wspace = 0.1, hspace=0.3)

# making a title for the figure
figure1.suptitle("Example : 4 Quantitative Vars Juxtaposed Scatter plots ")

plt.show()

#other option to try with subplot() , like frameon=False
```

#### We can use loop and make our code more compact and elegant

```python
#Doing the same as above using compact code and loop
figure1 = plt.figure(figsize = (10,5))

cols = d.columns
for i in [1,2,3,4]:
    subplt=  plt.subplot(2, 2,i)
    v=subplt.scatter(d["Month"],d.loc[:,cols[i]])
    subplt.set_title(cols[i])

# adjusting the space between the subplots
figure1.tight_layout(pad=3.0) 

# we can use the following too
#plt.subplots_adjust(wspace = 0.2, hspace=0.5)

# making a title for the figure
figure1.suptitle("Example : 4 Quantitative Vars Juxtaposed Scatter plots ")


plt.show()
```

#### Method 2: Using `plt.subplots()`

We can also do the same using `plt.subplots()`. Here also I'm using a loop to make the code more compact 

```python
#Doing the same as above using subplots
#figure1 = plt.figure(figsize = (10,5))
fig, subplts = plt.subplots(2, 2, figsize = (10,5)) #sharex='col', sharey='row', gridspec_kw={'hspace': 0, 'wspace': 0})

#notice the axes array "subplts" returned by .subplots() has shape similar to nrows x ncols
mysubplts =  subplts.flatten() # so we flatten tha array for easy access with one variable loop
cols = d.columns
for i in [1,2,3,4]:    
    v=mysubplts[i-1].scatter(d["Month"],d.loc[:,cols[i]])
    mysubplts[i-1].set_title(cols[i])

# adjusting the space between the subplots
fig.tight_layout(pad=3.0) 

# we can use the following too
#plt.subplots_adjust(wspace = 0.2, hspace=0.5)

# making a title for the figure
fig.suptitle("Example : 4 Quantitative Vars Juxtaposed Scatter plots ")

plt.show()
```
